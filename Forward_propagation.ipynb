```python
import pandas as pd

data = {
    'CGPA': [8, 7, 6, 9, 5, 8, 6, 9, 8, 6],
    'Marks': [8, 7, 6, 9, 5, 8, 6, 9, 8, 6],
    'SKILLS': [8, 8, 7, 8, 6, 8, 8, 9, 7, 6],
    'Exam': [9, 8, 4, 9, 4, 9, 4, 9, 4, 8],
    'Placed': [1, 1, 0, 1, 0, 1, 0, 1, 1, 0]
}

df = pd.DataFrame(data)

print(df)
X=df.iloc[:,:4]
X
```

```python
# Multi layer Perceptron

import numpy as np

W1=np.ones((4,3))
B1=np.ones((3,1))
W1
W2=np.ones((3,2))
B2=np.ones((2,1))
W2
W3=np.ones((2,1))
B3=np.ones((1,1))
W3
```

```python
k=X.iloc[2:].to_numpy().reshape(-1,1)
k
```

```python

B2.reshape(-1,1)
```

```python
for i in range(df.shape[0]):
  x_i = X.iloc[i, :].to_numpy().reshape(-1, 1)
  l=np.dot(W1.T,x_i)

  b=l+B1

  Activ=(1/(1+np.exp(-b)))
  print(Activ)
```

```python
for j in range(df.shape[0]):
  x_j = Activ.reshape(-1, 1)
  l1=np.dot(W2.T,x_j)

  b2=l1+B2
  Activ2=(1/(1+np.exp(-b2)))
  print(Activ2)
```

```python
for i in range(df.shape[0]):
  x_i=Activ2.reshape(-1,1)
  l2=np.dot(W3.T,x_i)
  b3=l2+B3
  Activ3=(1/(1+np.exp(-b3)))
  print(Activ3)
```
