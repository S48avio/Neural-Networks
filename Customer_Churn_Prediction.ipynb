```python
df=df.iloc[:,3:]
df
```

```python
df=pd.get_dummies(df,columns=['Geography','Gender'],drop_first=True)

df['Geography_Germany']=df['Geography_Germany'].replace(True,1)

df['Geography_Germany']=df['Geography_Germany'].replace(False,0)

df['Geography_Spain']=df['Geography_Spain'].replace(True,1)

df['Geography_Spain']=df['Geography_Spain'].replace(False,0)

df['Gender_Male']=df['Gender_Male'].replace(True,1)

df['Gender_Male']=df['Gender_Male'].replace(False,0)

```

```python
df
```

```python
X=df.drop(['Exited'],axis=1)
y=df['Exited']


X.shape
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(10,activation='relu',input_dim=11))
model.add(Dense(10,activation='relu'))

model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam',loss='binary_crossentropy')
```

```python
model.fit(X_train,Y_train,epochs=10)
```

```python
model.layers[0].get_weights()
```

```python
model.layers[1].get_weights()
```

```python
z=model.predict(X_test)
```

```python
import numpy as np
y_predi=np.where(z>0.5,1,0)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,y_predi)
```
